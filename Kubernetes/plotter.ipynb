```python
from matplotlib import pyplot as plt
%matplotlib inline

with open('httperf_output_clean.txt', 'r') as file:
    lines = file.readlines()

# Separate reply rate and session rate
reply_rates = [float(lines[i].strip()) for i in range(0, len(lines), 2)]
session_rates = [float(lines[i].strip()) for i in range(1, len(lines), 2)]

# Plot the graph
plt.plot(session_rates)
plt.xlabel('Session Rate')
plt.ylabel('Reply Rate')
plt.title('Reply Rate vs Session Rate')
plt.show()
```
