```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('log.csv')
x = data.iloc[:, 0]

y1 = data.iloc[:, 1]
y2 = data.iloc[:, 2]

fig, ax = plt.subplots(figsize=(10,6)) 

ax.plot(x, y1, label='CPU')
ax.plot(x, y2, label='MEM')

plt.xticks(rotation=45)

ax.set_xlabel('Utilization')
ax.set_ylabel('Time')
ax.legend()

plt.show()
```
